## Lambda Functions <br>
<p>Lambda functions... or "Anonymous Functions" are referring to inline functions with no name. The keyword lambda denotes the no name function, and executes within a single line. Without saving it to a variable; however, it is not able to be used, unless passed in either as a paramater or within list comprehension.<br>Written as "(keyword lambda) (one or more inputs) (colon) (function to be executed)"</p>

#### Syntax

```python
def add_two(x):
    return x + 2

add_two(4)
```

```python
# Lambda function SYNTAX
# Example of calling a Lambda function immediately

(lambda x: x+2)(4)
```

#### Saving to a Variable

```python
f_test = lambda x: x+2
f_test

f_test(4)
f_test(10)
```

#### Multiple Inputs

```python
# Multiple inputs wiht no variable

(lambda x,y,z: x *  y + z)(4,5,7)

mult_test = lambda x,y,z: x * y + z

mult_test(4,5,7)
```

#### Passing a Lambda into a Function

```python
def apply_func(f, name):
    return f(name)

apply_func(str.upper, 'brian')
```

```python
apply_func(lambda a_string: a_string[0], 'codingtemple')
```

#### Returning a Lambda from a Function

```python
# Regular defined function

def multiply_test(num):
    a = 5
    def inner(c,d):
        return (a * num) - (c * d)
    return inner


test = multiply_test(10)
print(test)
print(type(test))
```

```python
def return_lambda(num):
    a = 5
    return lambda c, d: (num * a) + (c * d)

lamb_test = return_lambda(15)
print(lamb_test)
print(type(lamb_test))
```

```python
lamb_test(4,5)
```

#### If Statements within Lambdas

```python
def do_something(num):
    if num > 10:
        return num*2
    else:
        return num+2
    
do_something(11)
```

```python
# Lambda input(s): return TRUE if (condition) else return FALSE

do_lambda = lambda num: num * 2 if num > 10 else num + 2

do_lambda(10)
```

## Map <br>
<p>The map function allows you to iterate over an entire list while running a function on each item of the list. This is why the map function works well with lambda's, because it simplifies things and you write less lines of code.<br>The syntax for a map function is "map(function to be used, list to be used)"<br>However, you must be careful, as the map function returns a map object, not a list. To turn it into a list we use the list() type conversion.</p>

#### Syntax

```python
help(map)
```

```python
# map(function, iterable -> list, Tuple, set, etc)

def squared_minus_two(a):
    return a**2 -2


numbers = [2,5,6,8,2,7,8,3,7]

mapped_nums = map(squared_minus_two, numbers)
mapped_nums_list = list(mapped_nums)
mapped_nums_list
```

```python
nums = [2,5,4,7,9]
more_nums = [3,4,6,1]



map(squared_minus_two, nums, more_nums)
```

#### Using Lambda's with Map

```python
mapped_squares_list = list(map(lambda x: x**2 -2, numbers))
mapped_squares_list
```

```python
list(map(lambda x,y: (x,y)))
```

## Filter() <br>
<p>Filter's are similar to the map function, where you're able to pass a function argument and a list argument and filter out something from the list based on the conditions passed. Similar to the map function, it returns a filter object, so you need to type convert it to a list()</p>

#### Syntax

```python
help(filter)
```

```python
names = ['Andy', 'Max', 'Evan', 'Angelica']

def starts_with_a(name):
    if name[0].lower() == 'a':
        return True
    else:
        return False
    
    
    
a_names_mapped = list(map(starts_with_a, names))

print(a_names_mapped)                      

a_names_filtered = list(filter(starts_with_a, names))

print(a_names_filtered)

starts_with_a('Arian')
```

#### Using Lambda's with Filter()

```python
a_names = list(filter(lambda n: True if n[0].lower() == 'a' else False, names))
a_names
```

```python
def remainder(x):
    return x % 3

remainder(10)

list(filter(remainder, [1,2,3,4,5,6,7,8,9]))

```

```python
list(map(remainder, [1,2,3,4,5,6,7,8,9]))
```

```python
def empty_or_full_list(x):
    if x**2 > 10:
        return [1,2,3]
    else:
        return []
    
list(filter(empty_or_full_list, [1,2,3,4,5,6]))
```

## Reduce() <br>
<p>Be very careful when using this function, as of Python 3 it's been moved to the 'functools' library and no longer is a built-in function.<br>The creator of Python himself, says to just use a for loop instead.</p>

#### Syntax

```python
from functools import reduce

help(reduce)
```

```python
list_1 = [2,4,6,8,10]

def add_nums(num1, num2):
    print("------------")
    print(f"num1: {num1}")
    print(f"num2: {num2}")
    return num1 + num2

result_add = reduce(add_nums, list_1)
print(result_add)
```

```python
list_1 = [2,4,6,8,10]

def sub_nums(num1, num2):
    print("------------")
    print(f"num1: {num1}")
    print(f"num2: {num2}")
    return num1 - num2

result_sub = reduce(sub_nums, list_1)
print(result_sub)
```

#### Using Lambda's with Reduce()

```python
result_lamb = reduce(lambda num1, num2: num1 + num2, list_1)
print(result_lamb)
```

## Recursion <br>
<p>Recursion means that a function is calling itself, so it contanstly executes until a base case is reached. It will then push the returning values back up the chain until the function is complete. A prime example of recursion is computing factorials... such that 5! (factorial) is 5*4*3*2*1 which equals 120.</p>

```python
def full_name(first, last):
    return first + ' ' + last

def greet_person(first, last, day_of_week):
    person_full_name = full_name(first, last)
    print(f"Hello {person_full_name}. It sure is beautiful on this {day_of_week}.")
    
greet_person('Jen', 'Dunlap', "Friday")
```

#### Implementing a Base Case

```python
def add_nums(num):
    # Set base case for recursive function
    if num <= 1:
        return num
    else:
        return num + add_nums(num-1)
    
add_nums(2)
```

#### Writing a Factorial Function

```python
# 5! = 5 * 4 * 3 * 2 * 1

def factorial(n):
    if n<=1:
        return 1
    else:
        return n * factorial(n-1)
    
factorial(5)
```

## Generators <br>
<p>Generators are a type of iterable, like lists or tuples. They do not allow indexing, but they can still be iterated through with for loops. They are created using functions and the yield statement.</p>

#### Yield Keyword <br>
<p>The yield keyword denotes a generator, it doesn't return so it won't leave the function and reset all variables in the function scope, instead it yields the number back to the caller.</p>

```python
def my_range(start, stop, step=1):
    while start < stop:
        yield start
        start += step
        
cool_range = my_range(0, 10)

for x in cool_range:
    print(x**2)
```

#### In-Class Exercise #6 <br>
<p>Create a generator that takes a number argument and yields that number squared, then prints each number squared until zero is reached.</p>

```python
def squared_generator(num):
    while num >= 0:
        yield num
        num = num -1
        
find_squared = squared_generator(10)

for x in find_squared:
    print(x**2)
```
